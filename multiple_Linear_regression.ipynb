#Encoding

```python
# df_State = pd.get_dummies(drop_first=True,df["State"])
df_State = pd.get_dummies(df["State"])
df_State
```

```python
dfDummies = pd.get_dummies(df["State"], prefix="State")
```

```python
dfDummies
```

```python
df = pd.concat([df,dfDummies], axis=1)

#axis=1 >>>>> mean concatenate columns
#axis=0 >>>>> mean concatenate rows
```

```python
df = df.drop(["State_Florida"], axis=1)
```

```python
df.head()
```

```python
df = df.drop(["State"], axis=1)
df.head()
```

#split data 

```python
x = df.drop("Profit", axis= 1) # dont affect the origin data as we didn't write inplace =true
y = df["Profit"]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 35)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#Build Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
model = lm.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

#Evaluate Model

```python
df = pd.DataFrame({"y_test" : y_test, "y_pred" 
                   : y_pred, "diff between y_pred and y_test" :abs(y_pred-y_test)})
df
```

```python
from sklearn.metrics import mean_absolute_error
MSE =mean_absolute_error(y_test, y_pred)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
model.score(X_train, y_train)
```
