```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
import seaborn as sns
```

#Bagging

```python
df_train=pd.read_csv('pendigits-tra.csv')
df_test=pd.read_csv('pendigits-tes.csv')
```

```python
# x_train,x_test,y_train,y_test=train_test_split()
x_train=df_train.iloc[:,:-1]
y_train=df_train.iloc[:,-1:]
x_test=df_test.iloc[:,:-1]
y_test=df_test.iloc[:,-1:]
```

```python
#-----------------Decesion_tree--------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

estimator = DecisionTreeClassifier(random_state=2022)
model=estimator.fit(x_train, y_train)
y_pred = estimator.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print(f'the accuracy of decesion tree = {acc}')
report = classification_report(y_test, y_pred)
print(report)
confusion_matrix(y_test,y_pred)
plot_confusion_matrix(model,x_test,y_test)
# plotEstimator(X_train, y_train, X_test, y_test, estimator, 'Decision Tree')
```

```python
#-----------bagging using decesion tree---------------

for nEst in range(1, 5):
  estimator = BaggingClassifier(n_estimators=nEst, random_state=2022)
  model=estimator.fit(x_train, y_train)
  y_pred = estimator.predict(x_test)
  acc = accuracy_score(y_test, y_pred)
  print(f'the accuracy of decesion tree = {acc}')
  #report = classification_report(y_test, y_pred)
  #print(report)
  #confusion_matrix(y_test,y_pred)
  plot_confusion_matrix(model,x_test,y_test)
```

```python
#-----------bagging using SVR ---------------

for nEst in range(1, 5):
  estimator = BaggingClassifier(base_estimator=SVC(),n_estimators=nEst, random_state=2022)
  model=estimator.fit(x_train, y_train)
  y_pred = estimator.predict(x_test)
  acc = accuracy_score(y_test, y_pred)
  print(f'the accuracy of decesion tree = {acc}')
  #report = classification_report(y_test, y_pred)
  #print(report)
  plot_confusion_matrix(model,x_test,y_test)
```

```python
#-----best number of estimator-----------

accuracies=[]
items=[20,50,90,120,180]
for nEst in items:
  estimator = BaggingClassifier(n_estimators=nEst, random_state=2022)
  model=estimator.fit(x_train, y_train)
  y_pred = estimator.predict(x_test)
  acc = accuracy_score(y_test, y_pred)
  accuracies.append(acc)
  print(f'the accuracy of decesion tree = {acc}')
  #report = classification_report(y_test, y_pred)
  #print(report)
  #confusion_matrix(y_test,y_pred)
  plot_confusion_matrix(model,x_test,y_test)
```

```python
print(accuracies)
```

```python

sns.barplot(x=items,y=accuracies).set(title='comparing number of estimators with thier accuracies',xlabel='number of estimators',ylabel='Accuracy')
```
